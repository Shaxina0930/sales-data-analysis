```python
smth=df[df["Product"]=="iPhone"]
```

```python
sales_by_city = smth['City'].value_counts().reset_index()
# sales_by_city.columns = ['City', 'smth']
```

```python
sales_by_city
```

```python
! pip install matplotlib pandas
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Grafik sozlamalari
plt.figure(figsize=(12, 6))  # Grafik o‘lchami

# Mahsulot A uchun chiziqli grafik
plt.plot(sales_by_city['City'],sales_by_city['count'],
         marker='o',
         label='Mahsulot A',
         color='#1f77b4',  # Zamonaviy ko‘k rang
         linewidth=2,
         markersize=8)



# Grafik elementlari
plt.title("Shaxarlar bo‘yicha mahsulotlar sotuvi", fontsize=16, pad=20, weight='bold')
plt.xlabel("Shaxarlar", fontsize=12)
plt.ylabel("Sotilgan birliklar soni", fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.legend(fontsize=11, loc='upper left')
plt.grid(True, linestyle='--', alpha=0.7, color='#d3d3d3')
plt.tight_layout()


# Grafikni ko‘rsatish
plt.show()
```
